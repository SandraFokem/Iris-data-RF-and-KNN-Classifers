```python
# Data preprocessing
from sklearn.model_selection import train_test_split

X = data[['Sepal Length', 'Sepal Width', 'Petal length', 'Petal width']] # these are the features, the independant variables
y = data['species'] # these are the dependable variables

X_train,X_test, y_train, y_test = train_test_split(X,y,test_size=0.33, random_state=5)
    #33% test and 67% training

```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=5)
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
# now lets test the algorythm, RF
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(n_estimators=5, n_jobs=-1, random_state=0, max_depth=2) 
model.fit(X_train, y_train) # training the classifier 
y_pred=model.predict(X_test)
accuracy= metrics.accuracy_score(y_test, y_pred)
print(accuracy)
```

```python
#feature importances for RF
feature_imp = pd.Series(model.feature_importances_,index=iris.feature_names).sort_values(ascending =False)
feature_imp  # we will use seaborn to visualize this because it provides additional plot types and a number of customized themes.
             #Matplotlib is just a superset of seaborn. 
```

```python
# now lets create a bar plot 
sns.barplot(x=feature_imp, y=feature_imp.index)
plt.xlabel('feature importance score')
plt.ylabel('features')
plt.title("Visualizing Important Features for Random Forest Classifier")
plt.legend()

#Result: the graph shows that sepal width is the least important feature for RandomForest
```

```python
# let's plot the accuracy against the number of trees for RF
n_estimators_settings= range(1, 7)
accuracy=[]
for n in n_estimators_settings:
    model=RandomForestClassifier(n_estimators=n)
    model.fit(X_train,y_train)
    accuracy.append(model.score(X_train, y_train))
    
print (accuracy)
plt.plot(n_estimators_settings, accuracy, label="acc on tr set")
plt.ylabel("accuracy")
plt.xlabel("n_estimators")
plt.legend()
```

```python
#now let's develop the KNN Classifier 
#and plot accuracy against number of n_neighbors
from sklearn.neighbors import KNeighborsClassifier 
k_range = list(range(1,35))
scores = []
for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X, y)
    y_pred = knn.predict(X)
    scores.append(metrics.accuracy_score(y, y_pred))
    
plt.plot(k_range, scores)
plt.xlabel('n_neighbors')
plt.ylabel('Accuracy Scores')
plt.title('Accuracy Scores for n_neighbors')
plt.show()
```

```python
#okay going to build some models classification and then look at its accuracy, recall precision

np.random.seed(1000)
# making a list of ml classification models i decided to add more classification models in to have more comparison such as svm, decision tree
models = []

def classification_Models(X_train,X_test, y_train, y_test ):
    

    
    models.append( ('RFC',  RandomForestClassifier()) )
    models.append( ('DTC',DecisionTreeClassifier()) )
    models.append( ('KNN', KNeighborsClassifier()) )
    models.append( ('SVM',  SVC()) )
# we used SVM and DTC because we wanted to create more models and compare it to our two classifiers of focus.
    modeloutcomes = []
    modelnames = []
    for name,model in models:
        v_results = cross_val_score(model, X_train, y_train, cv = 3, 
                                     scoring='accuracy', n_jobs = -1, verbose = 0)
        print(name,v_results.mean())
        modeloutcomes.append(v_results)
        modelnames.append(name)
        
    print(modeloutcomes)
    fig = plt.figure()
    ax = fig.add_subplot(111)
    ax.set_xticklabels(modelnames)
    plt.boxplot(modeloutcomes)
        
classification_Models(X_train,X_test, y_train, y_test)
```

```python
# Testing the accuracy, recall and precision, f1-score and support to decide which is the best model.
#f1-score is s the weighted average of Precision and Recall

for name,model in models:
    trainedmodel = model.fit(X_train,y_train)
    
    # predicting our models
    ypredict = trainedmodel.predict(X_test)
    
    acc = accuracy_score(y_test,ypredict)
    classreport = classification_report(y_test,ypredict)
    confMat = confusion_matrix(y_test,ypredict)
    #confusion matrix to describe the performance of a classification models on a set of our test data for which the true values are known
   #putting the nume/ title of each model alongside it's accuracy and prediction 
    print('\n**'+name)
    print('The accuracy: {}'.format(acc))
    print('The Classification Report:\n {}'.format(classreport))
    print('The Confusion Matrix:\n {}'.format(confMat))
    
#"support" is the count of the number of species     
    

    import pickle as pk
    #using this library to save my model
    
    with open('model_'+name+'.pickle','wb') as f:
        pk.dump(trainedmodel,f)
```
